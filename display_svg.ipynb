```python
import sys
import os
from PIL import Image
from matplotlib import pyplot as plt
%matplotlib inline
from cairosvg import svg2png

```

```python
base_directory = '/n/pfister_lab2/Lab/wisoo/DS2-final/data/cubicasa5k/high_quality_architectural/1191'

# base_directory = '/n/pfister_lab2/Lab/wisoo/cubicasa5k/high_quality_architectural/989'
```

```python
base_directory = base_directory + '/'
```

```python
file_names = os.listdir(base_directory)

for idx, file in enumerate(file_names):
    print(f"[{idx}]:  filename : {file}")
```

```python
index_ = 0
```

```python
img_name = file_names[index_]
```

```python
im = Image.open( base_directory +
    img_name)
plt.figure(figsize=(40,25))
plt.imshow(im)
```

```python
svg2png(
    url=base_directory+img_name, write_to=base_directory+"temp.png")

im = Image.open( base_directory +
    "temp.png")
plt.figure(figsize=(40,25))
plt.imshow(im)

# os.remove("temp.png")
```
